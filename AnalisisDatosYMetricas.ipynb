¿Cuándo es útil utilizar las métricas de distancia y cuándo las métricas de similitud?

Es util tomar las metricas de distancia en analitica de datos cuando se requiere saber en un termino absoluto la diferencia entre un par de puntos sin tener en cuenta criterios de comparacion, las metricas de similitud tienen utilidad cuando hay que comparar distancia entre varios objetos, arreglos, palabras.


¿Qué significa en la similitud coseno valores de 1, 0 y -1?

Un valor de 1 significa que los angulos de los vectores son identicos, apuntan a la misma direccion osea su angulo es 0 grados o uno es producto escalar del otro. Un valor de 0 significa que geometricamente no hay correlacion entre estos vectores, o son ortogonales y no tienen ninguna similitud. Y una similitud de -1 significa que el vector a forma un angulo de 180 grados con el vector B.




¿Cuáles son las aplicaciones que tendría utilizar la similitud coseno en la Analítica de Datos?

El uso que tiene en NLP donde cada palabra tiene un valor asignado y es un vector que se compara con otros vectores para determinar similitudes textuales, sugerencias de contenido similar que podria generar interacción basado en información existente.




¿Qué es la correlación y qué tipo de relación mide entre dos variables?

La correlacion es una variable que mide que tanto afectan los cambios de una variable en otra, toma en cuenta la covarianza de las variables a analizar y sus respecticas desviaciones estandar.




Si se agrega una constante a todos los valores de una de las variables, ¿cómo afecta esto al
coeficiente de correlación? Si se multiplican todos los valores por una constante positiva,
¿cómo se ve afectado?

Al sumar una constante a todos los elementos de un vector A, las diferencias respecto a la media son las mismas asi que la covarianza no cambia y la desviacion estandar tampoco, por tanto la correlacion es la misma que la del vector sin sumar.
Si se multiplica una constante por cada elemento, la media va a multiplicarse haciendo que la covarianza lo haga de manera proporcional, la desviacion estandar va a aumentar proporcionalmente, asi que se cancelan mutuamente en el calculo de la correlacion dando el mismo resultado.

```python
import numpy as np
from scipy.spatial import distance
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

def calcular_distancia_euclidiana(v1, v2):
    return distance.euclidean(v1, v2)

def calcular_similitud_coseno(v1, v2):
    return 1 - distance.cosine(v1, v2)

def calcular_similitud_jaccard(v1, v2):
    return 1 - distance.jaccard(v1 != 0, v2 != 0)

def calcular_correlacion(v1, v2):
    corr, _ = pearsonr(v1, v2)
    return corr

def calcular_distancia_hamming(v1, v2):
    return distance.hamming(v1, v2)

def calcular_distancia_l1(v1, v2):
    return distance.cityblock(v1, v2)

def calcular_distancia_supremum(v1, v2):
    return distance.chebyshev(v1, v2)


vec1 = np.array([1, 1,1,1])
vec2 = np.array([2,2,2,2])
vec3= np.array([0,1,0,1])
vec4=np.array([1,0,1,0])
vec5=np.array([0,-1,0,1])
vec6=np.array([1,0,-1,0])
vec7=np.array([1,1,0,1,0,1])
vec8=np.array([1,1,1,0,0,1])
vec9=np.array([2,-1,0,2,0,-3])
vec10=np.array([-1,1,-1,0,0,-1])

```

```python
print("a. X=vec1, Y=vec2")
print(f"   Coseno: {calcular_similitud_coseno(vec1, vec2):.4f}")
print(f"   Correlación: {calcular_correlacion(vec1, vec2):.4f}")
print(f"   Euclidiana: {calcular_distancia_euclidiana(vec1, vec2):.4f}")

print("\nb. X=vec3, Y=vec4")
print(f"   Coseno: {calcular_similitud_coseno(vec3, vec4):.4f}")
print(f"   Correlación: {calcular_correlacion(vec3, vec4):.4f}")
print(f"   Euclidiana: {calcular_distancia_euclidiana(vec3, vec4):.4f}")
print(f"   Jaccard: {calcular_similitud_jaccard(vec3, vec4):.4f}")
print(f"   Hamming: {calcular_distancia_hamming(vec3, vec4):.4f}")

print("\nc. X=vec5, Y=vec6")
print(f"   Coseno: {calcular_similitud_coseno(vec5, vec6):.4f}")
print(f"   Correlación: {calcular_correlacion(vec5, vec6):.4f}")
print(f"   Euclidiana: {calcular_distancia_euclidiana(vec5, vec6):.4f}")
print(f"   L1: {calcular_distancia_l1(vec5, vec6):.4f}")

print("\nd. X=vec7, Y=vec8")
print(f"   Coseno: {calcular_similitud_coseno(vec7, vec8):.4f}")
print(f"   Correlación: {calcular_correlacion(vec7, vec8):.4f}")
print(f"   Jaccard: {calcular_similitud_jaccard(vec7, vec8):.4f}")

print("\ne. X=vec9, Y=vec10")
print(f"   Coseno: {calcular_similitud_coseno(vec9, vec10):.4f}")
print(f"   Correlación: {calcular_correlacion(vec9, vec10):.4f}")
print(f"   Supremum (Lmax): {calcular_distancia_supremum(vec9, vec10):.4f}")
```

Arreglo Receta A: [0 1 2 3 4 5 6]

Arreglo Receta B: [7 8 2 0 4]

Arreglo Receta C: [ 0  8  2  9  3 10]

```python


def a_one_hot(indices, size=11):
    vector = np.zeros(size)
    vector[indices] = 1
    return vector


indices_a = [0, 1, 2, 3, 4, 5, 6]
indices_b = [7, 8, 2, 0, 4]
indices_c = [0, 8, 2, 9, 3, 10]


vec_a_hot = a_one_hot(indices_a)
vec_b_hot = a_one_hot(indices_b)
vec_c_hot = a_one_hot(indices_c)

print("Vector One-Hot Receta A:", vec_a_hot)
print("Vector One-Hot Receta B:", vec_b_hot)
print("Vector One-Hot Receta C:", vec_c_hot)

```

```python
print("Comparación Receta A vs Receta B:")
print(f"   Coseno: {calcular_similitud_coseno(vec_a_hot, vec_b_hot):.4f}")
print(f"   Jaccard: {calcular_similitud_jaccard(vec_a_hot, vec_b_hot):.4f}")
print(f"   Hamming: {calcular_distancia_hamming(vec_a_hot, vec_b_hot):.4f}")

print("\nComparación Receta A vs Receta C:")
print(f"   Coseno: {calcular_similitud_coseno(vec_a_hot, vec_c_hot):.4f}")
print(f"   Jaccard: {calcular_similitud_jaccard(vec_a_hot, vec_c_hot):.4f}")
print(f"   Hamming: {calcular_distancia_hamming(vec_a_hot, vec_c_hot):.4f}")

print("\nComparación Receta B vs Receta C:")
print(f"   Coseno: {calcular_similitud_coseno(vec_b_hot, vec_c_hot):.4f}")
print(f"   Jaccard: {calcular_similitud_jaccard(vec_b_hot, vec_c_hot):.4f}")
print(f"   Hamming: {calcular_distancia_hamming(vec_b_hot, vec_c_hot):.4f}")
```

La distancia de hamming cuenta cuantas posiciones son diferentes en un arreglo que debe ser de igual tamaño y cuando los ceros no significan algo relevante, se evidencia que es la metrica que mas varia respecto a los demas en los arreglos evaluados (0.5455,0.6364,0.4545). El coeficiente de jaccard solo tiene en cuenta el numero de coincidencias exitosas, es decir el numero de ingredientes de una receta que si estan en otra es decir que para una categorizacion acorde a preguntas como Si y No, solo formara una coeficiente ignorando lo que no sea coincidencia de Si.
