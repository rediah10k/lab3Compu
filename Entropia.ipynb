```python
import numpy as np
from scipy.stats import entropy

x = [0,1,1,0,1,0]
y = [0,1,0,0,1,1]


mis=mutual_info_score(x,y)
print(mis)


values_x, counts_x = np.unique(x, return_counts=True)
px = counts_x / counts_x.sum()

values_y, counts_y = np.unique(y, return_counts=True)
py = counts_y / counts_y.sum()


xy = list(zip(x, y))
values_xy, counts_xy = np.unique(xy, axis=0, return_counts=True)
pxy = counts_xy / counts_xy.sum()


Hx = entropy(px, base=np.e)
Hy = entropy(py, base=np.e)
Hxy = entropy(pxy, base=np.e)


Ixy = Hx + Hy - Hxy

print("p(x):", px)
print("p(y):", py)
print("p(x,y):", pxy)

print("H(X):", Hx)
print("H(Y):", Hy)
print("H(X,Y):", Hxy)
print("I(X,Y):", Ixy)
```
